```python
# Q1>
# Scatter Plot:
# Use: Scatter plots display the relationship between two continuous variables, 
# allowing you to visualize patterns or correlations.
# Seaborn function: seaborn.scatterplot()
# Line Plot:
# Use: Line plots are useful for visualizing trends or changes in data over time or any ordered variable.
# Seaborn function: seaborn.lineplot()
# Bar Plot:
# Use: Bar plots are suitable for displaying and comparing categorical data, 
# typically showing the distribution of a single variable or the relationship between two variables.
# Seaborn function: seaborn.barplot()
# Histogram:
# Use: Histograms provide a visual representation of the distribution 
# of a single variable, showing the frequency of values in different bins.
# Seaborn function: seaborn.histplot()
# Box Plot:
# Use: Box plots summarize the distribution of a continuous variable, displaying the median, 
# quartiles, and outliers, helping to identify skewness, variability, and potential outliers.
# Seaborn function: seaborn.boxplot()
```

```python
# Q2>
import seaborn as sns
import matplotlib.pyplot as plt

fmri=sns.load_dataset("fmri")
sns.lineplot(data=fmri,x="timepoint",y="signal",hue="event",style="region")
plt.show()
```

```python
# Q3>
import seaborn as sns
import matplotlib.pyplot as plt
titanic=sns.load_dataset("titanic")
sns.boxplot(data=titanic, x='pclass', y='age')
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.show()
```

```python
# Q4>
import seaborn as sns
import matplotlib.pyplot as plt
diamonds= sns.load_dataset("diamonds")
sns.histplot(data=diamonds, x='price', hue='cut')
plt.show()
```

```python
# Q5>
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
sns.pairplot(data=iris, hue='species')
plt.show()
```

```python
# Q6>
import seaborn as sns
import matplotlib.pyplot as plt
flights= sns.load_dataset("flights")
flights_pivot = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_pivot, cmap="YlGnBu")
plt.show()
```
