```python
# # Q1>
# Matplotlib is a data visualization library in Python that is used to create high-quality graphs, charts, and plots. 
# It provides a variety of tools and functions for creating different types of visualizations such as line plots, 
# scatter plots, histograms, bar charts, etc.

# Matplotlib is used to visualize and analyze data in fields such as statistics, finance, biology, engineering, and many others. 
# It provides a wide range of customization options that allow users to create publication-quality visualizations.

# Five plots that can be plotted using the Pyplot module of Matplotlib are:
# Line plot,Scatter plot,Histogram,Bar chart,Pie chart
```

```python
# Q2>
# A scatter plot is a type of plot used to visualize the relationship between two continuous variables. 
# It shows the values of two variables as points on a two-dimensional plane, 
# with the x-axis representing one variable and the y-axis representing the other.

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)

plt.title('Scatter plot of x and y')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# Q3>
# The subplot() function in Matplotlib is used to create multiple plots within a single figure.
# It allows you to create multiple subplots arranged in a grid, with each subplot displaying a different 
# set of data or a different type of plot. 

import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

plt.suptitle('Four line plots')

plt.show()
```

```python
# Q4>
# A bar plot is a type of plot in Matplotlib that displays data as rectangular bars. 
# It is commonly used to compare different categories of data and to show trends over time.

# Bar plots are useful when you want to compare different categories of data, 
# such as the performance of different companies, as in the given example.

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.title('Company profits')
plt.xlabel('Company')
plt.ylabel('Profit (in millions of dollars)')
plt.show()

plt.barh(company, profit)
plt.title('Company profits')
plt.xlabel('Profit (in millions of dollars)')
plt.ylabel('Company')
plt.show()
```

```python
# Q5>
# A box plot is a type of plot in Matplotlib that displays the distribution of a dataset. 
# It is commonly used to identify outliers and to compare the distribution of multiple datasets.

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

plt.boxplot(data)
plt.title('Box Plot')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.show()
```
